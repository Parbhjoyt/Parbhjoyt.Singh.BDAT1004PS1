3.Implement function triangleArea(a,b,c) that takes as input the lengths of the 3
sides of a triangle and returns the area of the triangle. By Heron's formula, the area 
of a triangle with side lengths a, b, and c is s(s - a)(s -b)(s -c), where 
s = (a+b+c)/2. 

```
/* CODE FOR QUESTION 3*/

static double triangleArea(double a,double b,double c)
{
    double s1 = a;
    double s2 = b;
    double s3 = c;
    double threeS = (s1 + s2 + s3) / 2;
    double threeArea = Math.Sqrt(threeS * (threeS - s1) * (threeS - s2) * (threeS - s3));
    return threeArea;
}

Console.WriteLine("Please input the lenght of three side of triangle ");
double threeA = Convert.ToDouble(Console.ReadLine());
double threeB = Convert.ToDouble(Console.ReadLine());
double threeC = Convert.ToDouble(Console.ReadLine());
Console.WriteLine("Area of the Triangle is : " + triangleArea(threeA, threeB, threeC));
```

Go to the editor

```
Console.WriteLine("Input the number of elements to be stored in the array :");
int fourIn = Convert.ToInt32(Console.ReadLine());
int[] fourArray = new int[fourIn];
Console.WriteLine("Input the elements");
for (int i = 0; i < fourIn; i++)
{
    fourArray[i] = Convert.ToInt32(Console.ReadLine());
}
int[] fourArrayEven = new int[fourIn];
int fourArrayEvenCounter = 0;
int[] fourArrayOdd = new int[fourIn];
int fourArrayOddCounter = 0;
for (int i = 0;i < fourIn; i++)
{
    if(fourArray[i] == 0) 
    { 
        fourArrayEven[fourArrayEvenCounter] = 0; 
        fourArrayEvenCounter+=1;
    }
    else if(fourArray[i]%2 == 0) 
    { 
        fourArrayEven[fourArrayEvenCounter] = fourArray[i];
        fourArrayEvenCounter+=1;
    }
    else
    {
        fourArrayOdd[fourArrayOddCounter] = fourArray[i];
        fourArrayOddCounter+=1;
    }
}
Console.WriteLine("Even Elements:");
for(int i = 0;i<fourArrayEvenCounter;i++)
{
    Console.WriteLine(fourArrayEven[i]);
}
Console.WriteLine("Odd Elements");
for(int j = 0;j < fourArrayOddCounter;j++)
{
    Console.WriteLine(fourArrayOdd[j]);
}
```
